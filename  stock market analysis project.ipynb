```python
import pandas as pd
from pandas_datareader import data,wb
from datetime import date
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
%matplotlib inline
```

```python
import yfinance as yf
start_date='1990-01-01'
end_date='2022-01-01'
symbols=['MSFT','AMZN','AAPL','GOOG','META']
stock=yf.download(symbols,start_date,end_date)
stock.tail()
```

```python
# Stock price
adj_date = stock["Adj Close"]
adj_date.plot()
plt.title("Tech Stocks Adjusted Price")
plt.xlabel("Date")
plt.ylabel("Adjusted Closing Price Over Time")
plt.grid(True)
plt.show()
```

```python
#stock price of Tech
fig = plt.figure(figsize=(20, 10))
ax1 = plt.subplot(2, 3, 1)
plt.plot(adj_date['MSFT'], color='blue')
plt.title('Microsoft')
plt.xlabel('Date')
plt.ylabel("Adj Close")
#Amazon
ax2 = plt.subplot(2, 3, 2)
plt.plot(adj_date['AMZN'], color='orange')
plt.title('Amazon')
plt.xlabel('Date')
plt.ylabel("Adj Close")
#Apple
ax3 = plt.subplot(2, 3, 3)
plt.plot(adj_date['AAPL'], color='green')
plt.title('Apple')
plt.xlabel('Date')
plt.ylabel("Adj Close")
#Google
ax4 = plt.subplot(2, 3, 4)
plt.plot(adj_date['GOOG'], color='red')
plt.title('Google')
plt.xlabel('Date')
plt.ylabel("Adj Close")
#Meta
ax5 = plt.subplot(2, 3, 5)
plt.plot(adj_date['META'], color='black')
plt.title('Meta')
plt.xlabel('Date')
plt.ylabel("Adj Close")
plt.show()
```

```python
# Simple rate of return
daily_return = adj_date.pct_change()
daily_return.plot()
plt.title("Daily Simple Rate of Return")
plt.xlabel("Date")
plt.ylabel("Rate of Return")
plt.grid(True)
plt.show()
```

```python
# Daily Return
fig = plt.figure(figsize=(20, 10))
ax1 = plt.subplot(2, 3, 1)
plt.plot(daily_return['MSFT'], color='green')
plt.title('Microsoft')
plt.xlabel('Date')
plt.ylabel('Daily Return')
#Amazon
ax2 = plt.subplot(2, 3, 2)
plt.plot(daily_return['AMZN'], color='red')
plt.title('Amazon')
plt.xlabel('Date')
plt.ylabel('Daily Return')
#Apple
ax3 = plt.subplot(2, 3, 3)
plt.plot(daily_return['AAPL'], color='orange')
plt.title('Apple')
plt.xlabel('Date')
plt.ylabel('Daily Return')
#Google
ax4 = plt.subplot(2, 3, 4)
plt.plot(daily_return['GOOG'], color='blue')
plt.title('Google')
plt.xlabel('Date')
plt.ylabel('Daily Return')
#Meta
ax5 = plt.subplot(2, 3, 5)
plt.plot(daily_return['META'], color='black')
plt.title('Meta')
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.show()
```

```python
# Mean rates of return
mean_returns = adj_date.pct_change().mean().plot(kind = 'bar', figsize = (8, 4))
plt.title("Mean rate of returns")
plt.xlabel("Symbols")
plt.ylabel("Mean Rate")
plt.show()
```

```python
# Variances of the returns
variance_returns = adj_date.var()
variance_returns.plot()
plt.title("Variances of the retuns")
plt.xlabel("Symbols")
plt.ylabel("Variances")
plt.grid(True)
plt.show()
```

```python
# Standard deviation of the returns
std_returns = daily_return.std()
std_returns.plot()
plt.title("Standard deviation of the returns")
plt.xlabel("Symbols")
plt.ylabel("Standard deviation")
plt.grid(True)
plt.show()
```

```python
# Correlation between the tech stock
import seaborn as sns
corr_data =adj_date.pct_change().corr(method='pearson')
sns.heatmap(corr_data, annot=True, cmap="summer")
plt.title('Correlation of stock return')
plt.show()  
```
